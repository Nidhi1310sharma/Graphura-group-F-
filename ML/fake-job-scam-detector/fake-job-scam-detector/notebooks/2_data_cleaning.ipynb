```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/job_data.csv")
print(df.shape)
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#Fill text columns
text_columns = [

    'title',
    'company',
    'skills',
    'description',
    'perks',
    'company_about',
    'company_website',
    'recruiter_email',
    'domain_name'

]

for col in text_columns:

    df[col] = df[col].fillna("")
```

```python
#Create Missing Information Features
df["has_description"] = (

    df["description"]

    .str.len()

    .gt(20)

    .astype(int)

)

df["has_skills"] = (

    df["skills"]

    .str.len()

    .gt(2)

    .astype(int)

)

df["has_company_website"] = (

    df["company_website"]

    .str.len()

    .gt(5)

    .astype(int)

)

df["has_recruiter_email"] = (

    df["recruiter_email"]

    .str.len()

    .gt(5)

    .astype(int)

)
```

```python
#Missing Info Score

#Higher = More Suspicious

df["missing_info_score"] = (

    4

    - (

        df["has_description"]

        + df["has_skills"]

        + df["has_company_website"]

        + df["has_recruiter_email"]

    )

)
```

```python
#Check result
df[

    [

        "title",

        "has_description",

        "has_skills",

        "has_company_website",

        "has_recruiter_email",

        "missing_info_score"

    ]

].head()
```

```python
#Check nulls again
df.isnull().sum()
```

```python
#Save cleaned data
df.to_csv(

    "../data/cleaned_data.csv",

    index=False

)

print("Saved Successfully")
```
