```python
import pandas as pd
```

```python
# LOAD FEATURE ENGINEERED DATASET
df = pd.read_csv(
    "../data/final_features.csv"
)
```

```python
# =====================================================
# GENERATE FRAUD REASONS
# =====================================================

def explain_job(row):

    reasons = []

    # NLP Signals

    if row["keyword_score"] > 0:
        reasons.append(
            "Scam keywords detected"
        )

    if row["contact_score"] > 0:
        reasons.append(
            "WhatsApp/Telegram detected"
        )

    if row["urgency_score"] > 0:
        reasons.append(
            "Urgent hiring language"
        )

    # Trust Signals

    if row["website_available"] == 0:
        reasons.append(
            "No company website"
        )

    if row["email_available"] == 0:
        reasons.append(
            "No recruiter email"
        )

    # Information Quality

    if row["short_description_flag"] == 1:
        reasons.append(
            "Very short description"
        )

    if row["skills_missing"] == 1:
        reasons.append(
            "No skills mentioned"
        )

    if row["company_about_missing"] == 1:
        reasons.append(
            "No company information"
        )

    if row["missing_info_score"] >= 2:
        reasons.append(
            "Missing critical information"
        )

    if len(reasons) == 0:
        reasons.append(
            "No major fraud indicators"
        )

    return " | ".join(reasons)
```

```python
# =====================================================
# CREATE FRAUD REASONS COLUMN
# =====================================================

df["fraud_reasons"] = df.apply(
    explain_job,
    axis=1
)

print("Fraud reasons generated")
```

```python
# =====================================================
# CREATE HUMAN READABLE RISK LEVEL
# =====================================================

def risk_level(score):

    if score >= 80:
        return "HIGH"

    elif score >= 50:
        return "MEDIUM"

    else:
        return "LOW"
```

```python
# =====================================================
# ASSIGN RISK LEVEL
# =====================================================

df["risk_level"] = df[
    "fraud_score_v2"
].apply(risk_level)
```

```python
# =====================================================
# TOP RISK JOBS
# =====================================================

df[
[
"title",
"company",
"fraud_score_v2",
"risk_level",
"fraud_reasons"
]
].sort_values(
    by="fraud_score_v2",
    ascending=False
).head(20)
```

```python
# =====================================================
# RISK LEVEL COUNTS
# =====================================================

df["risk_level"].value_counts()
```

```python
# =====================================================
# SAVE EXPLAINABLE DATASET
# =====================================================

df.to_csv(
    "../data/explainable_jobs.csv",
    index=False
)

print(
    "Saved: explainable_jobs.csv"
)
```
